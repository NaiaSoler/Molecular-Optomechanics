```python
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import Image
import qutip
from qutip import *
from tqdm import tqdm
from qutip.solver import correlation_2op_1t
from matplotlib import rc
from scipy.linalg import expm
from scipy.fft import ifftshift
import scipy.signal as signal
from scipy.special import factorial
import matplotlib as mpl
from matplotlib import cm
from brokenaxes import brokenaxes  # Librería para ejes rotos
import pandas as pd
from joblib import Parallel, delayed
from concurrent.futures import ProcessPoolExecutor
from optomechanics import *
```

```python
#SYSTEM'S PARAMETERS
hbar=1
kappa= 250e-3 
g0 = 0.003 
Omega= 0.02*kappa
wm=100e-3
wc = 2.5 
gamma_m= 1e-3
n_th=0.02 # for T=300k
wl = np.concatenate([np.linspace(wc+3*wm, wc+2*wm, 5), np.linspace(wc+2*wm, wc+wm, 5), 
                     np.linspace(wc+wm, wc, 5), np.linspace(wc, wc-wm, 5), np.linspace(wc-wm, wc-2*wm, 5)])
wlist=np.linspace(1, 2.5, 10000)
Delta= wc-wl
alpha= Omega/(kappa/2 + 1j*Delta) 
alpha2= alpha.conjugate()*alpha
g = g0*abs(alpha)
```

```python
data = []

for i in range(len(Delta)):
    Gamma_plus, Gamma_minus, Gamma_opt = Transition_rates(g0, alpha2[i], kappa, Delta[i], wm)
    nb = pop_phonons(gamma_m, Gamma_opt, n_th, Gamma_plus)
    S = [emission_spec(w, nb, Gamma_plus, Gamma_minus, Gamma_opt, gamma_m, wl[i], wm) for w in wlist]
    data.append({
        'g0': g0,
        'Δ': Delta[i],
        'Omega': Omega,
        'Gamma_plus': np.real(Gamma_plus),
        'Gamma_minus': np.real(Gamma_minus),
        'Gamma_opt': np.real(Gamma_opt),
        'Analytical S': np.real(S),
        'Analytical w': wlist,
        'C': g[i]**2/(kappa*gamma_m),
        'g/k': g[i]/kappa
    })

# Convertir a DataFrame
df = pd.DataFrame(data)
```

```python
#NUMERICAL SIMULATION
#Hamiltonian construction
Nc = 12  # Number of cavity states
Nm = 10 # Number of mech states

a = tensor(destroy(N=Nc), qeye(Nm))
b = tensor(qeye(Nc), destroy(N=Nm))
num_b = b.dag() * b
num_a = a.dag() * a

H = [hbar*Delta[i]*num_a+hbar*wm*num_b+1j*Omega*(a.dag()-a)-g0*num_a*(b.dag()+b) for i in range(len(Delta))]
c_ops = [np.sqrt(kappa)*a, np.sqrt((n_th+1)*gamma_m)*b, np.sqrt(n_th*gamma_m)*b.dag()]
```

```python
%%time

tlist = np.linspace(0, 30000, 50000)

results1 = Parallel(n_jobs=8)(delayed(spectrum_calculation)(H[i], tlist, wl[i], wm, c_ops, a, b) for i in tqdm(range(len(Delta))))
for key in results1[0].keys():
    df[key] = [res[key] for res in results1]
```

```python
#Name code: Omega_Hamiltonian
df.to_pickle("Results/Detuning_results/005_full.pkl")
```
