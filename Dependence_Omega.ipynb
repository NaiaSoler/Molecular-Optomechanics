```python
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import Image
import qutip
from qutip import *
from tqdm import tqdm
from qutip.solver import correlation_2op_1t
from matplotlib import rc
from scipy.linalg import expm
from scipy.fft import ifftshift
import scipy.signal as signal
from scipy.special import factorial
import matplotlib as mpl
from matplotlib import cm
from brokenaxes import brokenaxes  # Librería para ejes rotos
import pandas as pd
from joblib import Parallel, delayed
from concurrent.futures import ProcessPoolExecutor
from optomechanics import *
```

```python
#SYSTEM'S PARAMETERS
hbar=1
kappa= 250e-3 
g0 = 0.003
Omega= np.logspace(-2.6,-1, 10) 
wm=100e-3
wc = 2.5 
gamma_m= 1e-3
n_th= 0.02 # for T=300K: 0.02, for T=0K: 0
wl = wc
wlist=np.linspace(1, 2.5, 10000)
Delta= wc-wl
alpha= Omega/(kappa/2 + 1j*Delta) 
alpha2= alpha.conjugate()*alpha
g = g0*abs(alpha)
```

```python
T= 300 #ºC + 273.15 #K
freq=wm*(2*np.pi)/(4.135667696e-15) #rad/s
nth=1/(-1+np.exp(1.054571817e-34*freq/(1.380649e-23*T)))
print(nth)#(f"{nth:.3e}")
```

```python
data = []

for i in range(len(Omega)):
    Gamma_plus, Gamma_minus, Gamma_opt = Transition_rates(g0, alpha2[i], kappa, Delta, wm)
    nb = pop_phonons(gamma_m, Gamma_opt, n_th, Gamma_plus)
    S = [emission_spec(w, nb, Gamma_plus, Gamma_minus, Gamma_opt, gamma_m, wl, wm) for w in wlist]
    data.append({
        'g0': g0,
        'Δ': Delta,
        'Omega': Omega[i],
        'nth': n_th,
        'Gamma_plus': np.real(Gamma_plus),
        'Gamma_minus': np.real(Gamma_minus),
        'Gamma_opt': np.real(Gamma_opt),
        'Analytical S': np.real(S),
        'C': g[i]**2/(kappa*gamma_m),
        'g/k': g[i]/kappa
    })

# Convert to DataFrame
df = pd.DataFrame(data)
```

```python
#NUMERICAL SIMULATION
#Hamiltonian construction
Nc = 12  # Number of cavity states
Nm = 10 # Number of mech states

a = tensor(destroy(N=Nc), qeye(Nm))
b = tensor(qeye(Nc), destroy(N=Nm))
num_b = b.dag() * b
num_a = a.dag() * a
#Choose Hamiltonian
H = [hbar*Delta*num_a+hbar*wm*num_b+1j*Omega[i]*(a.dag()-a)-g0*num_a*(b.dag()+b) for i in range(len(Omega))]
Hlin = [hbar*wm*num_b+hbar*Delta*num_a-hbar*abs(alpha[i])*g0*(a.dag()+a)*(b.dag()+b) for i in range(len(Omega))]
c_ops = [np.sqrt(kappa)*a, np.sqrt((n_th+1)*gamma_m)*b, np.sqrt(n_th*gamma_m)*b.dag()]
```

```python
%%time

tlist = np.linspace(0, 30000, 50000)

results1 = Parallel(n_jobs=10)(delayed(spectrum_calculation)(H[i], tlist, wl, wm, c_ops, a, b) for i in tqdm(range(len(Omega))))
for key in results1[0].keys():
    df[key] = [res[key] for res in results1]
```
